```python
import nltk
import pandas as pd 
doc1=pd.read_csv("Example_Technical_Skills.csv")
doc=doc1['Technology Skills'].tolist()
doc3=pd.read_csv("Raw_Skills_Dataset.csv")
doc2=doc3['RAW DATA'].tolist()





found_skills =[]
stop_words = set(nltk.corpus.stopwords.words('english'))
 
filtered_tokens = [w for w in doc2 if w not in stop_words]

filtered_tokens = [w for w in doc2 if w.isalpha()]
 
bigrams_trigrams = list(map(' '.join, nltk.everygrams(filtered_tokens, 2, 3)))
 


for token in filtered_tokens:
    if token.lower() in doc:
        found_skills.add(token)

for ngram in bigrams_trigrams:
    if ngram.lower() in doc:
        found_skills.add(ngram)       


print(filtered_tokens)




```
